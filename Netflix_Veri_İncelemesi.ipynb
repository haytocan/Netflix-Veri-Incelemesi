#### Sayisal Degisken Tablosu

```python
def plot_hist(variable):
    plt.figure(figsize = (9,5)) #x ekseni için 9 birim ve yükseklik y ekseni için 5 birim belirledik.
    plt.hist(df[variable], bins = 10) # histogramı 10 eşit aralığa bolduk.
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} Hist Dagilimi".format(variable))
    plt.show()
```

```python
numericVar = ["Toplam Kutuphane Genisligi", "TV Gosteri Sayisi", "Film Sayisi"]
for n in numericVar:
    plot_hist(n)
```

### Plotly🔥

* **Plotly ile temel, standart ve premium paketlerin ulkelere gore fiyatlari gorsellestirildi**

```python
# Ulke - Temel Paket Aylik Maliyet ($)
fig = px.bar(df, x='Ulke', y='Aylik Maliyet - Temel ($)', color = "Aylik Maliyet - Temel ($)")
fig.show()
```

```python
# Ulke - Standart Paket Aylik Maliyet ($)
fig = px.pie(df, values='Aylik Maliyet - Standart ($)', names='Ulke')
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
fig.show()
```

```python
# Ulke - Premium Paket Aylik Maliyet ($)
fig = px.scatter(df, x="Ulke", y="Aylik Maliyet - Premium ($)")
fig.show()
```
